# Merged_match data

```python
# Only keep NameBirth and matchnum column in the dataframe
match_all = match_all.loc[:, ('NameBirth', 'matchnum')]
match_all
```

```python
full_list = match_all[match_all['NameBirth'] != 'nanu cornel99999999']
full_list.count()
```

```python
# sort by matchnum
actor_list = full_list.to_numpy()
actor_list90_name = []
actor_list90_val = []
for val in actor_list:
    name = val[0]
    match_num = val[1]
    
    if(not isinstance(match_num, str) and match_num > 86):
        actor_list90_name.append(name)
        actor_list90_val.append(match_num)
actor_list90_val
```

```python
# read node.csv
nodes = read_csv('UofT_nodes.csv', encoding = "UTF-8")
```

```python
# extract NameBirth from node.csv, generate a list of string
import csv
node_name_arr = []
with open("nodes_name.csv", newline='', encoding='UTF-8') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        node_name_arr.append(row['NameBirth'])
node_name_arr
```

```python
# Create a csv file to save output
header = ['NameBirth', 'matchnum']

with open('match_node.csv', 'a') as f:
    writer = csv.writer(f)
    writer.writerow(header)
```

```python
from fuzzywuzzy import process

for i in range (len(actor_list90_name)):
    extracted_name = actor_list90_name[i]
    print(extracted_name)
    val = process.extractOne(extracted_name, node_name_arr)
    name = val[0]
    match_num = val[1]
    print(name,match_num)
    with open('match_node.csv', 'a', newline='', encoding='UTF-8') as f:
        writer = csv.DictWriter(f, fieldnames=header)
        writer.writerow({'NameBirth': name, 'matchnum': match_num})
    
```
