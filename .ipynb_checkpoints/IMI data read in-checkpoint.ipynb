```python
# Define the function clean_text that takes in a string and output a string but with text cleaned.
def clean_text(text):
    if pd.notnull(text):
        tokens = text.split()
        table = str.maketrans("","", punctuation) #creates a one to one mapping of a character to its translation/replacement
        tokens = [word.translate(table) for word in tokens]
        tokens = [word for word in tokens if word.isalpha()]
        tokens = [word.lower() for word in tokens]
        tokens = [word for word in tokens if len(word) > 1]
        return ' '.join(tokens)
    else:
        return text
```

```python
# re-Define the target_name list with the Person's name cleaned
person_list = target_name[target_name.schema == 'Person']
target_name = person_list
target_name.iloc[:,2] = person_list.iloc[:,2].apply(clean_text)
# Also create a duplicate person for people with more than one birth-date in the table:
target_name = target_name.assign(birth_date = target_name['birth_date'].str.split(';')).explode('birth_date')
target_name['birth_date'] = target_name['birth_date'].fillna('NaN')
target_name.head()
```

```python
# also clean the nodes list:
nodes['NAME'] = nodes['NAME'].apply(clean_text)
nodes['BIRTH_DT'] = nodes['BIRTH_DT'].fillna('NaN')
nodes.head()
```

```python
# Transferring birth_date to 8 digit number, with NaN changed to 99999999 to reduce colinearity
def data_to_number(sanited_time):
    sp = sanited_time.split("-")
    time = "".join(sp)
    length = len(time)
    if sanited_time == 'NaN':
        num = '99999999'
    elif length < 8:
        d = 8 - length
        num = time + '0' *d
    else:
        num = time
    
    return num

def month_sanity_check(a):
    months = {
        'jan': '01',
        'feb': '02',
        'mar': '03',
        'apr': '04',
        'may': '05',
        'jun': '06',
        'jul': '07',
        'aug': '08',
        'sep': '09',
        'oct': '10',
        'nov': '11',
        'dec': '12'
        }
    try:
        sp = a.split("-")
        ez = months[sp[1][:3].lower()]
        sp[1] = ez
        sanited_time = "-".join(sp)
        return data_to_number(sanited_time)
    except:
        return data_to_number(a)
```

```python
# Apply the function month_sanity_check to Nodes csv
nodes['BIRTH_DT'] = nodes['BIRTH_DT'].apply(month_sanity_check)
nodes.head()
```

```python
#Apply the function month_sanity_check to Target dataframe
target_name['birth_date'] = target_name['birth_date'].apply(month_sanity_check)
target_name.head()
```

```python
# output the name+ birth_date: nodes_name.csv for later on fuzzywuzzy search:
import csv

header = ['NameBirth']
name_birth_list_nodes = nodes['NAME']+nodes['BIRTH_DT']

    
with open('nodes_name.csv', 'w', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)

    # write the header
    writer.writerow(header)

    # write multiple rows
    for x in name_birth_list_nodes.values.tolist():

        writer.writerow([x])
```

```python
# output the name+ birth_date: target_com.csv for later on fuzzywuzzy search:

header = ['NameBirth']
name_birth_list_target = target_name['name']+target_name['birth_date']
    
for c in map(chr,range(97, 123)):
    words = [i for i in name_birth_list_target if i.startswith(c)]
    with open(c+'.csv','w', encoding='UTF8', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(header)
        for x in words:
            writer.writerow([x])
```
