```python
from string import punctuation
```

```python
def clean_text(text):
    if pd.notnull(text):
        tokens = text.split()
        table = str.maketrans("","", punctuation) #creates a one to one mapping of a character to its translation/replacement
        tokens = [word.translate(table) for word in tokens]
        tokens = [word for word in tokens if word.isalpha()]
        tokens = [word.lower() for word in tokens]
        tokens = [word for word in tokens if len(word) > 1]
        return ' '.join(tokens)
    else:
        return text
```

```python
person_list['name'] = person_list.iloc[:,2].apply(clean_text)
```

```python
def data_to_number(sanited_time):
    sp = sanited_time.split("-")
    time = "".join(sp)
    length = len(time)
    if sanited_time == 'NaN':
        num = '00000000'
    elif length < 8:
        d = 8 - length
        num = time + '0' *d
    else:
        num = time
    
    return num

def month_sanity_check(a):
    months = {
        'jan': '01',
        'feb': '02',
        'mar': '03',
        'apr': '04',
        'may': '05',
        'jun': '06',
        'jul': '07',
        'aug': '08',
        'sep': '09',
        'oct': '10',
        'nov': '11',
        'dec': '12'
        }
    sp = a.split("-")
    try:
        ez = months[sp[1][:3].lower()]
        sp[1] = ez
        sanited_time = "-".join(sp)
        return data_to_number(sanited_time)
    except:
        return data_to_number(a)
```
