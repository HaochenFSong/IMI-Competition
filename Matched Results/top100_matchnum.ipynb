```python
#read the path
cwd = os.path.abspath('')
#list all the files from the directory
file_list = os.listdir(cwd)
file_list
```

```python
#list all csv files only
csv_files = glob.glob('*.{}'.format('csv'))
csv_files
```

```python
match_all = pd.DataFrame()
#append all files together
for file in csv_files:
            df_temp = pd.read_csv(file)
            match_all = match_all.append(df_temp, ignore_index=True)
match_all
```

# Merged_match data

```python
# Only keep NameBirth and matchnum column in the dataframe
match_all = match_all.loc[:, ('NameBirth', 'matchnum')]
match_all
```

```python
# delete duplicates
match_all.drop_duplicates(subset="NameBirth",
                     keep='first', inplace=True, ignore_index=False)
match_all
print(match_all)
```

```python
# sort by matchnum
match_all.sort_values(by='matchnum', inplace=False, ascending=False,ignore_index = True)

print(match_all)
```
